```python
import pandas as pd
import geopandas as gpd
```

```python
housing = pd.read_csv('data/Housing_New_York_Units_by_Building.csv')
housing = housing[['Project Name', 'Borough', 'Latitude', 'Longitude', 'Reporting Construction Type'
]]

housing['Phone Number'] = ""
housing['type'] = "Affordable Housing"
housing.additional_info = housing['Reporting Construction Type']
housing['url'] = ""
housing = housing[['Project Name', 'Latitude', 'Longitude', 'Phone Number', 'Borough', 'Reporting Construction Type', 'url']]
housing.columns=['name', 'latitude', 'longitude', 'phone_number','address', 'additional_info', 'url']
housing['type'] = 'Affordable Housing'

housing.head()
```

```python
grocery = pd.read_csv('data/NY.csv', encoding='latin-1')
grocery = grocery[grocery.CITY.isin(['NEW YORK ', 'BRONX'])]

grocery['additional_info'] = "Accepts SNAP/EBT"
grocery['phone_number'] = ""
grocery = grocery[['STORE_NAME', 'latitude', 'longitude', 'phone_number', 'ADDRESS', 'additional_info']]
grocery.columns=['name', 'latitude', 'longitude', 'phone_number','address', 'additional_info']
grocery['type'] = "Grocery Store"

grocery['url'] = ""

grocery.head()
```

```python
lib = gpd.read_file('data/Library.geojson')
lib['str_geom'] = lib['geometry'].apply(lambda x: str((x)).strip('0 POINT '))

lib['longitude'] = lib.str_geom.apply(lambda x: float(x.split(" ")[0].strip("(")))
lib['latitude'] = lib.str_geom.apply(lambda x: float(x.split(" ")[1].strip(")")))
```

```python
lib.head()
```

```python
lib['phone_number'] = ''
libraries = lib[['name', 'latitude', 'longitude', 'phone_number','streetname', 'system']]
libraries.columns=['name', 'latitude', 'longitude', 'phone_number', 'address', 'additional_info']
libraries['type'] = "Library"


libraries.head()
```

```python
for_mapbox = pd.DataFrame(columns=['name', 'latitude', 'longitude', 'phone_number', 'address', 'additional_info', 'type', 'url'])

for_mapbox = for_mapbox.append(housing)
for_mapbox = for_mapbox.append(libraries)
#for_mapbox = for_mapbox.append(schools)
for_mapbox = for_mapbox.append(grocery)

for_mapbox = for_mapbox.fillna('')
for_mapbox = for_mapbox[['name', 'latitude', 'longitude', 'phone_number','address', 'additional_info', 'type', 'url']]
```

```python
for_mapbox.head()
```

## Combining with Yelp

```python
yelp = pd.read_csv('data/nyc_yelp.csv')
yelp.rename(columns={'addtional_info':'additional_info',
                     'phone': 'phone_number'},inplace=True)
```

```python
yelp.columns
```

```python
for_mapbox_combined = for_mapbox.append(yelp)
```

```python
for_mapbox_combined.type = for_mapbox_combined.type.apply(lambda x: x.title())
for_mapbox_combined = for_mapbox_combined.reset_index()
for_mapbox_combined = for_mapbox_combined.drop(['Unnamed: 0', 'index'], axis=1)
```

```python
for_mapbox_combined.to_csv('for_mapbox_combined_nyc.csv')
```
